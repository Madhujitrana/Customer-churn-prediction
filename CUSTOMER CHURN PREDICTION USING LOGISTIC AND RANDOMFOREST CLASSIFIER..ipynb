```python
le=LabelEncoder()
```

```python
for i in df.select_dtypes(include="object"):
    new=le.fit_transform(df[i])
    df[i]=new
```

```python
x=df.drop(["Target"],axis=1)
y=df["Target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
from imblearn.over_sampling import SMOTE as smote
```

```python
sm=smote()
x_train_smote,y_train_smote=sm.fit_resample(x_train,y_train)
```

```python
scaler=StandardScaler()
```

```python
x_train_smote_scaled=scaler.fit_transform(x_train_smote)
```

```python
x_test_scaled=scaler.transform(x_test)
```

```python
log=logisticregression()
```

```python
log.fit(x_train_smote_scaled,y_train_smote)
```

```python
y_pred=log.predict(x_test_scaled)
```

```python
accuracy_score(y_test,y_pred)
```

```python
ran_class=random_class()
```

```python
ran_class.fit(x_train_smote_scaled,y_train_smote)
```

```python
r_y_pred=ran_class.predict(x_test_scaled)
```

```python
accuracy_score(y_test,r_y_pred)
```

```python
def function(a,b,c,d):
    log=logisticregression()
    log.fit(a,b)
    y_pred=log.predict(c)
    accu=accuracy_score(d,y_pred)
    accu_final=accu
    accu_name="logisticregression"
    
    if accu< 0.80: 
        
        log=random_class()
        log.fit(a,b)
        y_pred=log.predict(c)
        accu1=accuracy_score(d,y_pred)
        if accu1>accu:
            accu_final=accu1
            accu_name="Randomforestclassifier"
 
    if accu1<0.80:
        
        log=d_c()
        log.fit(a,b)
        y_pred=log.predict(c)
        accu2=accuracy_score(d,y_pred)
        if accu2>accu1:
            accu_final=accu2
            accu_name="decission_tree"
        
        
    
    return accu_final,accu_name
```

```python
print(function(x_train_smote_scaled,y_train_smote,x_test_scaled,y_test))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples required at a leaf node
    'max_features': ['auto', 'sqrt', 'log2']  # Number of features to consider for the best split
}
```

```python
reg=random_class()
```

```python
grid=gridsearchcv(estimator=reg, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid.fit(x_train_smote_scaled,y_train_smote)
```

```python
grid.best_params_
```

```python
reg=random_class(max_depth=30,
 max_features='auto',
 min_samples_leaf=4,
 min_samples_split= 2,
 n_estimators=200)
```

```python
reg.fit(x_train_smote_scaled,y_train_smote)
reg_y_pred=reg.predict(x_test_scaled)
accuracy_score(y_test,reg_y_pred)
```

```python
f1_score(y_test,reg_y_pred)
```
